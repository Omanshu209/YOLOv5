## **Loading & Model Architecture**

```python
import torch.hub as hub
```

```python
yolo = hub.load("ultralytics/yolov5", "yolov5s", pretrained = True)
yolo
```

## **Testing**

```python
def pred_yolo(link : str) -> None:
  pred = yolo(link)
  pred.print()
  pred.show()
```
