## **Loading The Trained Model & Making Predictions Using It!**

```python
yolo_trained = hub.load(
    "ultralytics/yolov5",
    "custom",
    path = "/content/yolov5/runs/train/exp/weights/last.pt"
)
```

```python
def pred_yolo(url : str) -> None:
  pred = yolo_trained(url)
  pred.print()
  pred.show()
```
